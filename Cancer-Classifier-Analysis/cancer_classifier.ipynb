## Introduction to the KNClassifier

<p>Letâ€™s begin by importing the breast cancer data from sklearn. We want to import the function load_breast_cancer from sklearn.datasets.</p>

```python
# Importing the libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Importing the dataset

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
#Loading teh dataset
breast_cancer_data = load_breast_cancer()
# print(breast_cancer_data.target)
# print(breast_cancer_data.target_names)

# Splitting the dataset
training_data, validation_data, training_labels, validation_labels = train_test_split(breast_cancer_data.data, breast_cancer_data.target, test_size=0.2,train_size=0.8, random_state=100)

# Creating the KN classifier
accuracies = []
for k in range(1,100):
  classifier = KNeighborsClassifier(n_neighbors=k)
  classifier.fit(training_data, training_labels)
  accuracies.append(classifier.score(validation_data,validation_labels))

# Plotting the results
import matplotlib.pyplot as plt
k_list = list(range(1,100))
plt.figure(figsize=(10,6))
plt.plot(k_list,accuracies,"b-",label="Model prediction")
plt.title("Breast Cancer Classifier Accuracy")
plt.legend()
plt.xlabel("K")
plt.ylabel("Validation Score")
plt.show()
plt.close()
```
